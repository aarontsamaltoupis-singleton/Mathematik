```python

# folge aller zahlen ohne 9
def c(n):
    c = []
    f = []
    x = 0
    while len(c)<n:
        if str(9) not in str(x):
            c.append(x)
            if x<=n:
                f.append(x)
                
            
        x+=1
    
    return [c,f]

# anzahl aller zahlen ohne 9 zwischen 1 und n
def f(n):
    f = []
    for x in range(1,n+1):
        if str(9) not in str(x):
            c.append(x)

    return f
    


# folge aller zahlen
def id(n):
    return list(range(n))
```

```python
print(c(10))
print(id(10))
```
